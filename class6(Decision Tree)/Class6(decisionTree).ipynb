**Wear Jacket or not?**

```python
df.drop("Days",axis=1,inplace=True)
```

```python
df.head()
```

```python
x = df.drop("Wear Jacket?",axis=1)
x.head()
```

```python
y = df[["Wear Jacket?"]]
y.head() # X's and Y's separetion
```

```python
#ML can only works with numerical value not catagorical value.so we need to do encoding
#Here we are doing lebelEncoding(Another way)
from sklearn.preprocessing import LabelEncoder
```

```python
le = LabelEncoder()
```

```python
from pandas.core.dtypes.common import is_numeric_dtype
```

```python
for col in x.columns:
  if is_numeric_dtype(x[col]):
    continue
  else:
    x[col]=le.fit_transform(x[col])
```

```python
x.head()
```

```python
df.head()#accending order
```

```python
#sunny>>>1 cloudy>>>0
#cold>>0 warm>>1
#indoor>>0 outdoor>>1
#from sklearn.tree import DecisionTreeClassifier (must import when DecisionTreeClassifier)
clf =DecisionTreeClassifier()
```

```python
clf.fit(x,y)
```

```python
clf.predict([[0,0,0]])
```

```python
clf.predict([[1,0,1]])
```
