```python
#Ordinal Encoder Using Loop
gender1=['female','male']
smoker1 = ['no', 'yes']
region1 = ['northeast', 'northwest', 'southeast', 'southwest']

cols=['gender','smoker','region']
category_dict = {
    'gender': gender1,
    'smoker': smoker1,
    'region': region1
}
for col in cols:
  oe2=OrdinalEncoder(categories=[category_dict[col]])
  df10[col]=oe2.fit_transform(df10[[col]])
```
