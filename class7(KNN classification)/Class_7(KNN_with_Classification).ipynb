```python
df.drop('id',axis=1,inplace=True)
```

```python
df.head()
```

```python
x = df.drop('diagnosis',axis=1)
x.head()
```

```python
y= df[['diagnosis']]
y.head()
```

```python
from sklearn.model_selection import train_test_split as tts
x_train,x_test,y_train,y_test =tts(x,y,test_size=0.3,random_state= 251)
```

```python
x_train.head()
```

# ***Applying KNN***

```python
from sklearn.neighbors import KNeighborsClassifier
ne = KNeighborsClassifier()
```

```python
ne.fit(x_train,y_train)
```

```python
pred = ne.predict(x_test)
pred
```

```python
ne.predict(x_test)
```

# ***Evaluate Model***

```python
from sklearn.metrics import accuracy_score,confusion_matrix,roc_curve,classification_report,RocCurveDisplay
```

```python
print(classification_report(y_test,pred))
```

```python
con_matrix = confusion_matrix(y_test,pred)
```

```python
con_matrix
```

```python
import seaborn as sns
```

```python
plt.figure(figsize=(12,8))
sns.heatmap(con_matrix,annot= True)
plt.xlabel('Predicted Result')
plt.ylabel('Actual value')
plt.savefig("heatmap.png")
```

```python
accuracy_score(y_test,pred)
```
